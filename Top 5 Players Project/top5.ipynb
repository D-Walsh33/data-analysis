```python
# Import needed libraries:
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read in data
df = pd.read_csv('./data/top5-players.csv')
df
```

```python
usa_players = df[df['Nation'] == 'us USA']
usa_players
fw_players = df[df['Pos'] == 'FW']
fw_players
prem = df[df['Comp'] == 'eng Premier League']
bund = df[df['Comp'] == 'de Bundesliga']
```

```python
sns.catplot(data=prem, x='Pos', y='Gls_90', hue='Squad')
plt.savefig('./images/catplot.png')
```

```python
sns.scatterplot(
    data=prem,
    y='Gls_90',
    x='xG_90',
    hue='Squad'
)
plt.savefig('./images/premScatter.png')
```

```python
sns.scatterplot(
    data=bund,
    x='xG_90',
    y='Gls_90',
    hue='Squad'
)
plt.savefig('./images/bundScatter.png')
```

```python
sns.lineplot(
    data=prem,
    x='Age',
    y='Gls_90'
)
plt.savefig('./images/premLineAge.png')
```

```python
sns.relplot(
    data=df,
    x='xG_90',
    y='Gls_90',
    hue='Pos',
    col="Comp"
)
plt.savefig('./images/topScatter.png')
```

```python
sns.histplot(
    data=prem,
    x='xG_90',
    hue='Pos',
    multiple='stack'
)
```

```python
sns.kdeplot(data=prem, x='xG_90', hue='Pos')
```

```python
sns.displot(
    data=df,
    x='Age',
    col='Comp',
    kind='kde',
    rug=True
)
```
