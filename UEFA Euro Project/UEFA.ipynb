```python
# Import the libraries needed
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read players csv
players = pd.read_csv('./data/euro_lineups.csv', parse_dates=['birth_date'])
players
```

```python
# Find all England players
england_players = players[players['country_code'] == 'ENG']
england_players
```

```python
# Max number of games played for England national team
england_players['name'].value_counts().max()
```

```python
# Find average height and weight all different national teams where height and weight are listed
height_weight = players[players['height'].notna() & players['weight'].notna()]
avgHW = height_weight.groupby('country_code')[['weight', 'height']].mean()
avgHW.sort_values('height', ascending=False)['height'].plot(kind='bar')
```

```python
# Top 10 players with most starts on the field
players[players['start'] == 'field'].groupby('name')['start'].value_counts().sort_values(ascending=False).head(10)
```

```python
# group by games

starters2020 = players[(players['start'] == 'field') & (players['year'] == 2020)]
starters2020
```

```python
# What match had the shortest starting squad
shortest = starters2020[['id_match','country_code', 'name', 'height']]
shortest.fillna(180).groupby(['id_match', 'country_code'])['height'].mean().sort_values().head(1)

```

## Austria had the smallest squad when they played the North Macedonia.

```python
game2024442 = players[(players['id_match'] == 2024442) &(players['start'] == 'field')][['country_code','name_shirt', 'start_position_x', 'start_position_y']]

game2024442_AUT = game2024442[game2024442['country_code'] == 'AUT']
x_AUT = np.array(game2024442_AUT['start_position_x']) 
y_AUT = np.array(game2024442_AUT['start_position_y'])

game2024442_MKD = game2024442[game2024442['country_code'] == 'MKD']
x_MKD = np.array(game2024442_MKD['start_position_x']) 
y_MKD = np.array(game2024442_MKD['start_position_y'])

plt.figure(1)
plt.subplot(211)
scatter = plt.scatter(x_MKD,y_MKD)
ax = scatter.axes
ax.invert_yaxis()
plt.subplot(212)
plt.scatter(x_AUT, y_AUT)




```

```python
def formations(id_match, dataframe):
    """Returns a plot of the formations for a given match"""
    
    df = dataframe[dataframe['id_match'] == id_match]
    starters = df[df['start'] == 'field']
    country_code1 = starters['country_code'].sort_values().head(1).values[0]
    country_code2 = starters['country_code'].sort_values().tail(1).values[0]
    team1 = starters[starters['country_code'] == country_code1]
    x1 = np.array(team1['start_position_x'])
    y1 = np.array(team1['start_position_y'])
    team2 = starters[starters['country_code'] == country_code2]
    x2 = np.array(team2['start_position_x'])
    y2 = np.array(team2['start_position_y'])

    plt.figure(1)
    plt.subplot(211)
    scatter = plt.scatter(x1,y1)
    ax = scatter.axes
    ax.invert_yaxis()
    plt.subplot(212)
    plt.scatter(x2, y2)
    return
```

```python
formations(2024442, players)
```
